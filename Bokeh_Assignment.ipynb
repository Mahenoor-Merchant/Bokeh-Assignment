## ***Q1. How can you create a Bokeh plot using Python code?***

```python
#first import the following:
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
#importing Data
from bokeh.plotting import figure,output_file,show
from bokeh.sampledata.iris import flowers
output_file('output.html')
```

```python
#creating a Bokeh plot
p=figure(title="flower")
p.xaxis.axis_label="petal length"
p.xaxis.axis_label="sepal length"
p.circle(flowers['petal_length'],flowers['sepal_length'])
show (p)
```

## **Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.**

Glyphs are nothing but visual shapes that are drawn to represent the data such as circles, squares, lines, rectangles, etc.  

```python
from bokeh.plotting import figure, show
from bokeh.io import output_file

# prepare some data
x = [1, 2, 4, 6, 7]
y = [7, 6, 3, 9, 10]

# create a new plot with figure function
p = figure(title="Circle Glyph", plot_width=450, plot_height=400)

# create circle glyph
p.circle(x=x, y=y, size=25, fill_color="red")
p.circle_cross(x=[2, 4, 6, 8], y=[5, 8, 9, 11], size=25,
			fill_color="blue", fill_alpha=0.3, line_width=2)
p.circle_x(x=[4, 7, 2, 6], y=[7, 2, 4, 9], size=25,
		fill_color="green", fill_alpha=0.6, line_width=2)

# show the results
output_file('circle.html')
show(p)
```

## **Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?**

Bokeh is a Python library that allows you to create interactive visualizations for web browsers. Bokeh provides a server component called Bokeh Server, which enables you to build and deploy interactive plots and applications that can be updated in real time.

The Bokeh Server allows you to create dynamic and interactive visualizations by combining the power of Python on the server-side and JavaScript on the client-side. With Bokeh Server, you can generate plots and applications that respond to user interactions, such as zooming, panning, hovering, or clicking.

```python
from bokeh.models import HoverTool

x=[2,3,4,5,6]
y=[3,4,5,6,7]
output_file('plot.html')


p=figure(title="line")
p.xaxis.axis_label="X"
p.yaxis.axis_label="y"
p.scatter(x,y,fill_color='purple',legend_label = "data points",size=20)

hover = HoverTool(tooltips=[("x", "@x"), ("y", "@y")])
p.add_tools(hover)
show (p)
```
